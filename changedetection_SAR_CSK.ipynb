```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os

# Define the folder path where H5 files are located
folder_path = 'C:/Users/amusu/OneDrive/Documents/Changedetection/SAR'

# Define the dataset name
dataset_name = 'S01/QLK'

# Function to read a dataset from an H5 file
def read_h5_dataset(file_path, dataset_name):
    with h5py.File(file_path, 'r') as h5_file:
        # Check if the dataset exists
        if dataset_name in h5_file:
            # Read the dataset and return it
            dataset = h5_file[dataset_name][:]
            return dataset
        else:
            print(f"Dataset '{dataset_name}' not found in file: {file_path}")
            return None

# Get a list of all H5 files in the folder
h5_files = [file for file in os.listdir(folder_path) if file.endswith('.h5')]

# Sort the files to process them in order (optional, depending on naming convention)
h5_files.sort()

# Loop through the H5 files and process them in pairs
for i in range(len(h5_files) - 1):
    # Get file paths for the current pair of files
    file1_path = os.path.join(folder_path, h5_files[i])
    file2_path = os.path.join(folder_path, h5_files[i + 1])
    
    # Read the QLK datasets from both files
    dataset1 = read_h5_dataset(file1_path, dataset_name)
    dataset2 = read_h5_dataset(file2_path, dataset_name)
    
    # Check if both datasets were successfully read
    if dataset1 is not None and dataset2 is not None:
        # Calculate the difference between the two datasets
        difference = dataset1 - dataset2
        
        # Create a figure with three subplots
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))
        
        # Plot the first dataset
        ax1 = axes[0]
        img1 = ax1.imshow(dataset1, cmap='viridis')
        ax1.set_title(f'QLK from {h5_files[i]}')
        plt.colorbar(img1, ax=ax1)
        
        # Plot the second dataset
        ax2 = axes[1]
        img2 = ax2.imshow(dataset2, cmap='viridis')
        ax2.set_title(f'QLK from {h5_files[i + 1]}')
        plt.colorbar(img2, ax=ax2)
        
        # Plot the difference between the two datasets
        ax3 = axes[2]
        img3 = ax3.imshow(difference, cmap='viridis')
        ax3.set_title('Difference')
        plt.colorbar(img3, ax=ax3)
        
        # Display the plots
        plt.tight_layout()
        plt.show()
    else:
        print(f"Could not read QLK datasets from both files: {h5_files[i]} and {h5_files[i + 1]}")
```
